```python
import numpy as np
```

```python
skills = np.array([
    "Python",
    "NumPy",
    "Pandas",
    "Matplotlib",
    "Machine Learning",
    "Deep Learning",
    "NLP",
    "Computer Vision",
    "AWS",
    "GenAI"
])
```

```python
["Python","NumPy","Pandas"]
```

```python
progress = np.array([
    30,
    10,
    5,
    5,
    5,
    0,
    0,
    0,
    0,
    0
])
```

```python
print("Skills")
print(skills)

print("\nProgress")
print(progress)
```

```python
average_progress = np.mean(progress)

print("\nAverage Progress")
print(average_progress)
```

```python
highest_index = np.argmax(progress)

print(highest_index)
```

```python
strongest_skill = skills[highest_index]

print("\nStrongest Skill")
print(strongest_skill)
```

```python
lowest_index = np.argmin(progress)
weakest_skill = skills[lowest_index]
```

```python
remaining = 100 - progress

print("\nRemaining Learning")
print(remaining)
```

```python
print("\n========== SUMMARY ==========")

print("Average Progress:", average_progress)

print("Strongest Skill:", strongest_skill)

print("Focus Area:", weakest_skill)

print("\nRemaining Learning Percentage:")

for i in range(len(skills)):
    print(skills[i], ":", remaining[i], "% remaining")
```
